```python
# simple clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
```

```python
# encoding the mfr column
label_encoder=LabelEncoder()
df['mfr_encoded']= label_encoder.fit_transform(df.mfr)
df['mfr_encoded'].unique()
```

```python
# scaling numeric columns
cols_for_clustering= ['calories','protein','fat','carbo','sodium','fiber','potass','sugars','mfr_encoded']
scaler= StandardScaler()
scaled_data= scaler.fit_transform(df[cols_for_clustering])
scaled_data[1]
```

```python
# perform K means clustering
kmeans= KMeans(n_clusters=3, random_state= 42)
clusters= kmeans.fit_predict(scaled_data)
```

```python
df['clusters']= clusters
```

```python
# plot
plt.figure(figsize=(10,8))

for cluster in df.clusters.unique():
    plt.scatter(df.loc[df.clusters==cluster, 'mfr'],
                df.loc[df.clusters==cluster,'protein'],
                label= f'Cluster {cluster}', alpha= 0.9)
plt.xlabel('Manufacturer')
plt.ylabel('Protein')
plt.title('Clusters of cereals(Manufacturer VS Protein)')
plt.legend()
plt.show()
```

```python
# get nutritional statistics of the clusters
for cluster in df.clusters.unique():
    print(f'Cluster: {cluster}')
    print(df[df.clusters==cluster].describe())
    print('----')
```
